```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("cleaned_Dataset.csv")
```

```python
df_new = pd.DataFrame(data = df, columns = ["anne_sektor","baba_sektor"])
```

```python
df_new = df_new.replace(['0', '-'], 'Unknown')

# Or alternatively, use NaN and fill later:
# data = data.replace(['0', '-'], pd.NA)
```

```python
df_new
```

```python
from sklearn.preprocessing import LabelEncoder

# Apply label encoding
label_encoder = LabelEncoder()

# Encoding 'anne_sektor' and 'baba_sektor' columns
df_new['anne_sektor_encoded'] = label_encoder.fit_transform(df_new['anne_sektor'])
df_new['baba_sektor_encoded'] = label_encoder.fit_transform(df_new['baba_sektor'])
```

```python
df_new = pd.get_dummies(df_new, columns=['anne_sektor', 'baba_sektor'], prefix=['anne', 'baba'], dummy_na=True)
```

```python
df_new = df_new.drop(["anne_nan","baba_nan"],axis =1)
df_new
```

```python
df=df.drop(["anne_sektor","baba_sektor"],axis =1)
```

```python
cleanedData = pd.concat([df,df_new],axis = 1)
cleanedData
```

```python
cleanedData['cinsiyet'] = cleanedData['cinsiyet'].map({'erkek': 0, 'kadin': 1 ,"Belirtmek istemiyorum":-1})
```

```python
cleanedData
```

```python
cleanedData.to_csv("cleanedData.csv")
```

```python
from datetime import datetime
```

```python
cleanedData['dogum_tarihi'] = pd.to_datetime(cleanedData['dogum_tarihi'])

# Extract the year of birth
cleanedData['birth_year'] = cleanedData['dogum_tarihi'].dt.year
cleanedData['age'] = 2024 - cleanedData['birth_year']
cleanedData["age"]
```

```python
cleanedData = cleanedData.drop(["birth_year","dogum_tarihi"],axis =1)
```

```python
cleanedData.to_csv("cleanedData.csv")
```
